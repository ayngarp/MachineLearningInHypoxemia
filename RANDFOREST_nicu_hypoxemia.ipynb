```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('nicu_val_all.csv')
xdat = data.iloc[:, 1:-1].values
ydat = data.iloc[:, -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators = 20, random_state = 0)
reg.fit(xdat, ydat)
```

```python
reg.predict([[60]])
```

```python
x_grid = np.arange(min(xdat), max(xdat), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter (xdat, ydat, color = 'black')
plt.plot(x_grid, reg.predict(x_grid), color = 'red')
plt.title('The Correlation between the Respiration Rate of a Neonate to their Oxygen Saturation to assess the risk of Hypoxemia (RF)')
plt.xlabel('Respiration Rate (breaths per min)')
plt.ylabel('Oxygen Saturation (%)')
plt.show()
```
